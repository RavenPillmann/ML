# Hot Dog / Not Hot Dog

A now trite example, this notebook is meant to explore CNNs by building my own convolutional neural network using Keras to detect whether an image is a hot dog or not. The data was collected from this Kaggle set (https://www.kaggle.com/dansbecker/hot-dog-not-hot-dog/home), which in turn is extracted from the Food 101 dataset (https://www.kaggle.com/dansbecker/food-101/home). This is purely for educational purposes, which I hope goes without saying.

## Goal

To create a hot dog/not hot dog predictor without using transfer learning.

## Exploring the Data

```python
import numpy as np
from glob import glob
import random
random.seed(100)

# Load the training and testing data
training_hot_dog_files = np.array(glob("data/train/hot_dog/*"))
training_not_hot_dog_files = np.array(glob("data/train/not_hot_dog/*"))
testing_hot_dog_files = np.array(glob("data/test/hot_dog/*"))
testing_not_hot_dog_files = np.array(glob("data/test/not_hot_dog/*"))

print('There are %s training images of hot dogs.' % (len(training_hot_dog_files)))
print('There are %s training images of not hot dogs.' % (len(training_not_hot_dog_files)))
print('There are %s testing images of hot dogs.' % (len(testing_hot_dog_files)))
print('There are %s testing images of not hot dogs.' % (len(testing_not_hot_dog_files)))
```

There are an equal number of training and testing hot dog and not hot dog images in this data set. If we were to take a random image and unconditionally say it was a hot dog, we would have a 50% chance of being right. Therefore, I'm setting 50% to be the benchmark that I want to beat.

Here's a quick look at a hot dog and a not hot dog image.

```python
import cv2
import matplotlib.pyplot as plt                        
%matplotlib inline  

img = cv2.imread(training_hot_dog_files[0])
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img2 = cv2.imread(training_not_hot_dog_files[0])
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
plt.imshow(img2_rgb)
plt.show()
```

Delicious.

So now we probably want to start looking at converting these images to tensors. Tensors are required when running keras with tensorflow as the backend, so we'll write a few functions to convert file paths into tensors. These images are in color, so we'll make tensors that have three layers. These functions are coming from the Udacity Dog Project repo (https://github.com/udacity/dog-project).

## Making Tensors

```python
from keras.preprocessing import image                  
from tqdm import tqdm # For adding a neat progress bar

# Convert into a 224 by 224 image, have each tensor be of shape (undefined, 224, 224, 3)
def path_to_tensor(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    return np.expand_dims(x, axis=0)

# Get a (n, 224, 224, 3) tensor
def paths_to_tensor(img_paths):
    print('starting to convert these images')
    list_of_tensors = [path_to_tensor(img_path) for img_path in tqdm(img_paths)]
    print('converted all images for these paths')
    return np.vstack(list_of_tensors)
```

The images are already separated into training and testing sets, and hot dog and not hot dog images. We'll need to do a couple of things before we train and test. Specifically, we'll need to shuffle our training data and take out a validation set.

```python
from keras.utils import np_utils

# Create training targets/inputs
training_targets = np_utils.to_categorical(np.append(np.ones(len(training_hot_dog_files), dtype=int),
                                                     np.zeros(len(training_not_hot_dog_files), dtype=int)), 2)
training_inputs = np.append(training_hot_dog_files, training_not_hot_dog_files)

# Create testing targets/inputs
testing_targets = np.append(np.ones(len(testing_hot_dog_files), dtype=int),
                                                    np.zeros(len(testing_not_hot_dog_files), dtype=int))
testing_inputs = np.append(testing_hot_dog_files, testing_not_hot_dog_files)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(training_inputs, training_targets, random_state=453, shuffle=True, test_size=0.25)
```

Here I split the training targets and inputs into training and validation sets. The validation set is 25% of the original training set, which is arbitrarily chosen. Now we can begin turning out images into tensors.

```python
from PIL import ImageFile                            
ImageFile.LOAD_TRUNCATED_IMAGES = True

train_tensors = paths_to_tensor(X_train).astype('float32')/255
valid_tensors = paths_to_tensor(X_valid).astype('float32')/255
test_tensors = paths_to_tensor(testing_inputs).astype('float32')/255
```

Sweet, so now we have tensors that we can feed into a convolutional neural network. With that in mind, we'll start the process of training and building a CNN. I'll start with something pretty basic to see what initial performance is like.

## First Model

```python
from keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D
from keras.layers import Dropout, Flatten, Dense
from keras.models import Sequential

model = Sequential()

model.add(Conv2D(input_shape=(224, 224, 3), filters=8, kernel_size=2))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(Conv2D(filters = 16, kernel_size=2))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(Conv2D(filters=32, kernel_size=2))
model.add(MaxPooling2D(pool_size=2, strides=2))
model.add(GlobalAveragePooling2D())
model.add(Dense(2, activation='softmax'))

model.summary()
```

In this initial model, we're simply adding more and more filters and reducing the height and width for each sequential layers of the neural network. I will compile this and check performance against the validation set. I will use 'rmsprop' as the optimizer of this first model just by default.

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint  

checkpointer = ModelCheckpoint(filepath='hot_dog_saved_models/model1.best.hdf5', verbose=1, save_best_only=True)
model.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=30, batch_size=10,
          callbacks=[checkpointer], verbose=1)
```

This first model did alright, but with a loss of 0.6597 and validation loss of 0.6465, it seems like there may be a bit of underfitting occuring here. I should try adding more layers to improve this models performance.

## Second Model (more layers)

In this second model, I will add a third convolutional layer with 64 filters, as well as another dense layer at the end.

```python
model2 = Sequential()

model2.add(Conv2D(input_shape=(224, 224, 3), filters=8, kernel_size=2))
model2.add(MaxPooling2D(pool_size=2, strides=2))
model2.add(Conv2D(filters = 16, kernel_size=2))
model2.add(MaxPooling2D(pool_size=2, strides=2))
model2.add(Conv2D(filters=32, kernel_size=2))
model2.add(MaxPooling2D(pool_size=2, strides=2))
model2.add(Conv2D(filters=64, kernel_size=2))
model2.add(MaxPooling2D(pool_size=2, strides=2))
model2.add(GlobalAveragePooling2D())
model2.add(Dense(64, activation='relu'))
model2.add(Dense(2, activation='softmax'))

model2.summary()
```

With roughly 8 times the number of connections as the first model, this model will take a while to train without a GPU. I will increase the batch size to offset some of this.

```python
model2.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpointer2 = ModelCheckpoint(filepath='hot_dog_saved_models/model2.best.hdf5', verbose=1, save_best_only=True)
model2.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=30, batch_size=30,
          callbacks=[checkpointer2], verbose=1)
```

There isn't much of an improvement in terms of loss with these changes. The second model's validation loss, at it's best, was still 0.62296, compared to 0.64039 with the first model. However, things are moving in the right direction. With that in mind, I will continue to add layers to a third model.

## Model 3 (more layers still)

```python
model3 = Sequential()

model3.add(Conv2D(input_shape=(224, 224, 3), filters=8, kernel_size=2))
model3.add(MaxPooling2D(pool_size=2, strides=2))
model3.add(Conv2D(filters = 16, kernel_size=2))
model3.add(MaxPooling2D(pool_size=2, strides=2))
model3.add(Conv2D(filters=32, kernel_size=2))
model3.add(MaxPooling2D(pool_size=2, strides=2))
model3.add(Conv2D(filters=64, kernel_size=2))
model3.add(MaxPooling2D(pool_size=2, strides=2))
model3.add(Conv2D(filters=128, kernel_size=2))
model3.add(MaxPooling2D(pool_size=2, strides=2))
model3.add(GlobalAveragePooling2D())
model3.add(Dense(64, activation='relu'))
model3.add(Dense(32, activation='relu'))
model3.add(Dense(2, activation='softmax'))

model3.summary()
```

```python
model3.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpointer3 = ModelCheckpoint(filepath='hot_dog_saved_models/model3.best.hdf5', verbose=1, save_best_only=True)
model3.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=30, batch_size=30,
          callbacks=[checkpointer3], verbose=1)
```

Sweet. Continuing the trend, adding another convolutional layer and dense layer improved the best validation loss to 0.61484. The corresponding training loss is 0.636, which indicates to me that this model may still be underfitting. I will try with more epochs and see if any improvements arise. 

## Model 3.1 (Model 3 with more epochs)

```python
checkpointer3_1 = ModelCheckpoint(filepath='hot_dog_saved_models/model3_1.best.hdf5', verbose=1, save_best_only=True)
model3.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=50, batch_size=30,
          callbacks=[checkpointer3_1], verbose=1)
```

So we plateaue a little at a validation loss of 0.61116. After that it looked like a hefty amount of overfitting began to occur. For this next model I want to mitigate some of this by adding some dropout layers. 

## Model 4 (dropout layers)

```python
model4 = Sequential()

model4.add(Conv2D(input_shape=(224, 224, 3), filters=8, kernel_size=2))
model4.add(Dropout(0.2))
model4.add(MaxPooling2D(pool_size=2, strides=2))
model4.add(Conv2D(filters = 16, kernel_size=2))
model4.add(Dropout(0.2))
model4.add(MaxPooling2D(pool_size=2, strides=2))
model4.add(Conv2D(filters=32, kernel_size=2))
model4.add(Dropout(0.2))
model4.add(MaxPooling2D(pool_size=2, strides=2))
model4.add(Conv2D(filters=64, kernel_size=2))
model4.add(Dropout(0.2))
model4.add(MaxPooling2D(pool_size=2, strides=2))
model4.add(Conv2D(filters=128, kernel_size=2))
model4.add(Dropout(0.2))
model4.add(MaxPooling2D(pool_size=2, strides=2))
model4.add(GlobalAveragePooling2D())
model4.add(Dense(64, activation='relu'))
model4.add(Dropout(0.2))
model4.add(Dense(32, activation='relu'))
model4.add(Dropout(0.2))
model4.add(Dense(2, activation='softmax'))

model4.summary()
```

```python
model4.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpointer4 = ModelCheckpoint(filepath='hot_dog_saved_models/model4.best.hdf5', verbose=1, save_best_only=True)
model4.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=50, batch_size=30,
          callbacks=[checkpointer4], verbose=1)
```

We seem to not achieve an improvement with this model. Up to this point the activation models have been linear, and I'm interested in seeing if adding relu activation functions helps performance at all.

## Model 5 (Relu activation function, different optimizer)

```python
model5 = Sequential()

model5.add(Conv2D(input_shape=(224, 224, 3), filters=8, kernel_size=2, activation='relu'))
model5.add(MaxPooling2D(pool_size=2, strides=2))
model5.add(Conv2D(filters = 16, kernel_size=2, activation='relu'))
model5.add(MaxPooling2D(pool_size=2, strides=2))
model5.add(Conv2D(filters=32, kernel_size=2, activation='relu'))
model5.add(MaxPooling2D(pool_size=2, strides=2))
model5.add(Conv2D(filters=64, kernel_size=2, activation='relu'))
model5.add(MaxPooling2D(pool_size=2, strides=2))
model5.add(Conv2D(filters=128, kernel_size=2, activation='relu'))
model5.add(MaxPooling2D(pool_size=2, strides=2))
model5.add(GlobalAveragePooling2D())
model5.add(Dense(64, activation='relu'))
model5.add(Dense(32, activation='relu'))
model5.add(Dense(2, activation='softmax'))

model5.summary()
```

```python
model5.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpointer5 = ModelCheckpoint(filepath='hot_dog_saved_models/model5.best.hdf5', verbose=1, save_best_only=True)
model5.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=50, batch_size=30,
            callbacks=[checkpointer5], verbose=1)
```

Sweet, it looks like using Adam as well as ReLU activation functions improves our models performance, and we achieved a validation loss of 0.57522 with our best parameters. Next, I'm interested in seeing whether using 'same' padding in our Max Pooling layers improves performance. During the Udacity Machine Learning Nanodegree lesson on padding, it was mentioned that 'same' is generally believed to perform better than 'valid'.

## Model 6 (Using 'same' padding)

```python
model6 = Sequential()

model6.add(Conv2D(input_shape=(224, 224, 3), filters=8, kernel_size=2, activation='relu'))
model6.add(MaxPooling2D(pool_size=2, strides=2, padding='same'))
model6.add(Conv2D(filters = 16, kernel_size=2, activation='relu'))
model6.add(MaxPooling2D(pool_size=2, strides=2, padding='same'))
model6.add(Conv2D(filters=32, kernel_size=2, activation='relu'))
model6.add(MaxPooling2D(pool_size=2, strides=2, padding='same'))
model6.add(Conv2D(filters=64, kernel_size=2, activation='relu'))
model6.add(MaxPooling2D(pool_size=2, strides=2, padding='same'))
model6.add(Conv2D(filters=128, kernel_size=2, activation='relu'))
model6.add(MaxPooling2D(pool_size=2, strides=2, padding='same'))
model6.add(GlobalAveragePooling2D())
model6.add(Dense(64, activation='relu'))
model6.add(Dense(32, activation='relu'))
model6.add(Dense(2, activation='softmax'))

model6.summary()
```

```python
model6.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpointer6 = ModelCheckpoint(filepath='hot_dog_saved_models/model6.best.hdf5', verbose=1, save_best_only=True)
model6.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=50, batch_size=30,
            callbacks=[checkpointer6], verbose=1)
```

Validation loss with this model is slightly higher after 50 epochs than the previous model. For the next model, I am inspired by the VGG-16 model architecture, in particular by how the layers follow a pattern of two convolutions, one max pooling. I am also going to use an SGD optimizer with momentum to avoid getting stuck in local minima.

## Model 7 (Doubling up on convolution layers)

```python
model7 = Sequential()

model7.add(Conv2D(input_shape=(224, 224, 3), filters=8, kernel_size=2, activation='relu'))
model7.add(Conv2D(filters=8, kernel_size=2, activation='relu'))
model7.add(MaxPooling2D(pool_size=2, strides=2))
model7.add(Conv2D(filters = 16, kernel_size=2, activation='relu'))
model7.add(Conv2D(filters = 16, kernel_size=2, activation='relu'))
model7.add(MaxPooling2D(pool_size=2, strides=2))
model7.add(Conv2D(filters=32, kernel_size=2, activation='relu'))
model7.add(Conv2D(filters=32, kernel_size=2, activation='relu'))
model7.add(MaxPooling2D(pool_size=2, strides=2))
model7.add(Conv2D(filters=64, kernel_size=2, activation='relu'))
model7.add(Conv2D(filters=64, kernel_size=2, activation='relu'))
model7.add(MaxPooling2D(pool_size=2, strides=2))
model7.add(Conv2D(filters=128, kernel_size=2, activation='relu'))
model7.add(Conv2D(filters=128, kernel_size=2, activation='relu'))
model7.add(MaxPooling2D(pool_size=2, strides=2))
model7.add(GlobalAveragePooling2D())
model7.add(Dense(64, activation='relu'))
model7.add(Dense(32, activation='relu'))
model7.add(Dense(2, activation='softmax'))

model7.summary()
```

```python
from keras import optimizers

model7.compile(optimizer=optimizers.SGD(momentum=0.9), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpointer7 = ModelCheckpoint(filepath='hot_dog_saved_models/model7.best.hdf5', verbose=1, save_best_only=True)
model7.fit(train_tensors, y_train, validation_data=(valid_tensors, y_valid), epochs=50, batch_size=30,
            callbacks=[checkpointer7], verbose=1)
```

The approach to adding more parameters to this model didn't seem to help matters. In fact, it seems that the most performant model so far is model 5. 

```python
model5.load_weights('hot_dog_saved_models/model5.best.hdf5')

test_predictions = [np.argmax(prediction) for prediction in model5.predict(test_tensors)]
correct_guesses = np.sum(np.equal(np.array(test_predictions), np.array(testing_targets)))
test_accuracy = float(correct_guesses) / len(test_predictions)
print('The accuracy on the test set is %s' % (test_accuracy))
```

Well, I've hit my basic benchmark of 50% accuracy. Not too thrilled about 58% accuracy, but it's better than randomly guessing and considering this is just for practice, I'm pretty alright with it. I want to try training with transfer learning and see how that works, though. It might also be a good exercise to try and improving this model for generalization by augmenting the data.
